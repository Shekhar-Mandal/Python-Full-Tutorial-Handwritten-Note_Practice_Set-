```python
# Q2. Accept the gender from the user as char and print the 
#  respective greeting message
#  Ex - Good Morning Sir (on the basis of gender)

gender=input("Enter your gender as M and F :").upper()


if (gender=='M'):
    print(f'Good Morning Sir')
elif (gender=='F'):
    print(f'Good Morning madam')
else:
    print("please fill your gender in F and M , male and female")

```

```python
# Q3. Accept an integer and check whether it is an even number 
#  or odd.
number=int(input("Enter the Number : "))

if (number%2==0):
    print(f"Hey the number is even : {number}")
else:
    print(f"Hey the number is Odd : {number}")
```
